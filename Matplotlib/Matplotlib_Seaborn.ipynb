```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Cargar el archivo CSV
df = pd.read_csv("data/superstore_dataset2012.csv")

# 2. Ver las primeras filas para confirmar la carga
print(df.head())
```

```python
# ----------------------------
# 1. Grfico univariante con Matplotlib (Histograma de Ventas)
# ----------------------------
plt.figure(figsize=(8,5))
plt.hist(df["Sales"], bins=30, color="skyblue", edgecolor="black")
plt.title("Distribucin de Ventas (Sales)", fontsize=14)
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.grid(axis="y", alpha=0.4)
plt.show()
# Conclusin: La mayora de las ventas son pequeas (<1000), pero existen algunos valores atpicos muy altos.
```

```python
import seaborn as sns
# ----------------------------
# 2. Grfico univariante con Seaborn (Boxplot de Ganancias)
# ----------------------------
plt.figure(figsize=(7,5))
sns.boxplot(x=df["Profit"], color="lightgreen")
plt.title("Distribucin de Ganancias (Profit)", fontsize=14)
plt.xlabel("Ganancias")
plt.show()
# Conclusin: La mayora de las ganancias estn entre 0 y 500, pero hay prdidas (valores negativos) y algunos outliers muy altos.

```

```python
# ----------------------------
# 3. Grfico bivariante con Matplotlib (Dispersin Ventas vs Ganancias)
# ----------------------------
plt.figure(figsize=(8,6))
plt.scatter(df["Sales"], df["Profit"], alpha=0.5, color="purple")
plt.title("Relacin entre Ventas y Ganancias", fontsize=14)
plt.xlabel("Ventas")
plt.ylabel("Ganancias")
plt.grid(alpha=0.3)
plt.show()
# Conclusin: No todas las ventas elevadas generan ganancias, incluso algunas altas producen prdidas.

```

```python
# ----------------------------
# 4. Grfico bivariante con Seaborn (Dispersin coloreada por Categora)
# ----------------------------
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Sales", y="Profit", hue="Category", alpha=0.7, palette="Set2")
plt.title("Ventas vs Ganancias segn Categora", fontsize=14)
plt.xlabel("Ventas")
plt.ylabel("Ganancias")
plt.show()
# Conclusin: Se observa que Tecnologa tiende a concentrarse en mayores ventas, mientras que Muebles tiene ms casos con prdidas.

```

```python
# ----------------------------
# 5. Grfico multivariante con Seaborn (Heatmap de correlacin)
# ----------------------------
plt.figure(figsize=(8,6))
corr = df[["Sales","Quantity","Discount","Profit"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Mapa de calor de correlaciones", fontsize=14)
plt.show()
# Conclusin: Existe correlacin negativa entre Descuento y Ganancias, lo cual indica que mayores descuentos tienden a reducir la rentabilidad.
```

```python
# ----------------------------
# 6. Subplots organizados
# ----------------------------
fig, axes = plt.subplots(2,2, figsize=(12,10))

# Histograma
axes[0,0].hist(df["Sales"], bins=30, color="skyblue", edgecolor="black")
axes[0,0].set_title("Histograma de Ventas")

# Boxplot Profit
sns.boxplot(x=df["Profit"], color="lightgreen", ax=axes[0,1])
axes[0,1].set_title("Boxplot de Ganancias")

# Dispersin Sales vs Profit
axes[1,0].scatter(df["Sales"], df["Profit"], alpha=0.5, color="purple")
axes[1,0].set_title("Ventas vs Ganancias")

# Heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, ax=axes[1,1])
axes[1,1].set_title("Correlaciones")

plt.tight_layout()
plt.show()
```

```python
# ----------------------------
# 7. Guardar una figura como imagen
# ----------------------------
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Sales", y="Profit", hue="Category", alpha=0.7, palette="Set1")
plt.title("Ventas vs Ganancias por Categora")
plt.xlabel("Ventas")
plt.ylabel("Ganancias")
plt.savefig("data/ventas_ganancias_categoria.png")
plt.close()
```
